```python
import nltk
nltk.download('words')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('treebank')
```

```python
from nltk import tokenize
para = "Hello. My name is Shibu. Today you'll be learning NLTK."
sents = tokenize.sent_tokenize(para)
sents
```

```python
sent = tokenize.word_tokenize(sents[1])
sent
```

```python
from nltk import tag
tagged_sent = tag.pos_tag(sent)
tagged_sent
```

```python
from nltk import chunk
tree = chunk.ne_chunk(tagged_sent)
tree
```

```python
sent2 = tokenize.word_tokenize(sents[2])
sent2
```

```python
from nltk import tag
tagged_sent = tag.pos_tag(sent2)
tagged_sent
```

```python
from nltk import chunk
tree2 = chunk.ne_chunk(tagged_sent)
tree2
```

## Multilingual corpora

```python
# Indian Language POS Tagged Corpus includes samples of Indian text annotated with part-of-speech tags
# Show words from each text

for text in indian.fileids():
  print(text, indian.words(text)[:10])
```

```python
# The Brown corpus

import nltk
nltk.download('brown')
nltk.download('universal_tagset')

from nltk.corpus import brown
```
