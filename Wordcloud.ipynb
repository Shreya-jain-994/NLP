```python
from goose3 import Goose
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import warnings
from wordcloud import WordCloud
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# object to call Goose()
# extracting context from the data

g = Goose()
```

```python
url = "https://en.wikipedia.org/wiki/Data_science"
```

```python
# extracting url
article = g.extract(url)
```

```python
article
```

```python
# cleaned text article
article.cleaned_text
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
# argument
# converting article into tokens because no of words appearing
tokens = nltk.tokenize.word_tokenize(article.cleaned_text)

# fetch and display first 10 elements in the list
tokens[:10]
```

```python
# frequency of each word
frequency = nltk.FreqDist(tokens)
frequency.most_common(10)
```

```python
# printing most common words
most_common = frequency.most_common(10)
most_common
```

```python
# Frequency Distribution Plot
# True means ascending
# False means descending

import matplotlib.pyplot as plt
frequency.plot(10,cumulative=False)
plt.show()
```

```python
# object is cloud method is WordCloud()
# cloud is data
cloud = WordCloud()

# generating most common words and display as cloud representation
cloud = cloud.generate(article.cleaned_text)
```

```python
# The bigger the word means the more number of times the word is appearing in the dataset
# like word data is appearing 57 times, so it's size is large

plt.imshow(cloud)
```
